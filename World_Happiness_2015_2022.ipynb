**Happiness** is a key indicator of well-being and quality of life across the world. To understand global happiness trends, we are analyzing the **World Happiness Report (WHR)** dataset from **2015 to 2022**. This dataset provides valuable insights into how happiness has changed over time, which countries have improved or declined, and the overall trends in global well-being.

For our analysis, we are focusing on four key variables:

**Country** – The name of the country being analyzed.

**Year** – The year in which the data was recorded.

**Happiness Rank** – The ranking of a country based on its happiness score.

**Happiness Score** – A numerical score representing the happiness level of a country.

```python
# Import necessary libraries and mount Google Drive
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

data_path = "/content/drive/MyDrive/Colab/csv_files"

#Loading csv files
df_2015 = pd.read_csv(f"{data_path}/2015.csv")
df_2016 = pd.read_csv(f"{data_path}/2016.csv")
df_2017 = pd.read_csv(f"{data_path}/2017.csv")
df_2018 = pd.read_csv(f"{data_path}/2018.csv")
df_2019 = pd.read_csv(f"{data_path}/2019.csv")
df_2020 = pd.read_csv(f"{data_path}/2020.csv")
df_2021 = pd.read_csv(f"{data_path}/2021.csv")
df_2022 = pd.read_csv(f"{data_path}/2022.csv")

# Print all dataframes
print(df_2015.columns)
print(df_2016.columns)
print(df_2017.columns)
print(df_2018.columns)
print(df_2019.columns)
print(df_2020.columns)
print(df_2021.columns)
print(df_2022.columns)
```

```python
# Rename columns to make them consistent across all years
df_2015 = df_2015.rename(columns={
    "Happiness Rank": "Happiness Rank",
    "Happiness Score": "Happiness Score",
    "Country": "Country"
})

df_2016 = df_2016.rename(columns={
    "Happiness Rank": "Happiness Rank",
    "Happiness Score": "Happiness Score",
    "Country": "Country"
})

df_2017 = df_2017.rename(columns={
    "Happiness.Rank": "Happiness Rank",
    "Happiness.Score": "Happiness Score",
    "Country": "Country"
})

df_2018 = df_2018.rename(columns={
    "Overall rank": "Happiness Rank",
    "Score": "Happiness Score",
    "Country or region": "Country"
})

df_2019 = df_2019.rename(columns={
    "Overall rank": "Happiness Rank",
    "Score": "Happiness Score",
    "Country name": "Country"
})

df_2020 = df_2020.rename(columns={
    "Overall rank": "Happiness Rank",
    "Score": "Happiness Score",
    "Country name": "Country"
})

df_2021 = df_2021.rename(columns={
    "Country name": "Country",
    "Ladder score": "Happiness Score",
    "Overall rank": "Happiness Rank"
})

df_2022 = df_2022.rename(columns={
    "RANK": "Happiness Rank",
    "Happiness score": "Happiness Score",
    "Country": "Country"
})

print(df_2015.columns)
print(df_2016.columns)
print(df_2017.columns)
print(df_2018.columns)
print(df_2019.columns)
print(df_2020.columns)
print(df_2021.columns)
print(df_2022.columns)
```

```python
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019
df_2020['Year'] = 2020
df_2021['Year'] = 2021
df_2022['Year'] = 2022

# Concatenate all yearly DataFrames into a single DataFrame
df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022], ignore_index=True)

# Printing shape and head
print(df.head())
print(df.shape)
```

```python
# Select relevant columns for analysis
whr = df[['Country', 'Happiness Rank', 'Happiness Score', 'Year']]

# Display the first few rows, dataset shape, and count of missing values
print(whr.head())
print(whr.shape)
print(whr.isnull().sum())
```

```python
# Handle missing values and convert data types
whr.loc[:, 'Happiness Rank'] = whr['Happiness Rank'].fillna(whr['Happiness Rank'].mean())
whr.loc[:, 'Happiness Score'] = pd.to_numeric(whr['Happiness Score'], errors='coerce')
whr.loc[:, 'Happiness Score'] = whr['Happiness Score'].fillna(whr['Happiness Score'].mean())
whr['Country'] = whr['Country'].ffill()

# Display updated DataFrame
print(whr.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**1. Happiness Score Over Time for Different Countries/Regions**

```python
# Get the top 10 happiest countries in 2022
top_10_countries = whr[whr['Year'] == 2022].nsmallest(10, 'Happiness Rank')['Country']

# Filter data for these top 10 countries across all years
top_10_data = whr[whr['Country'].isin(top_10_countries)]

plt.figure(figsize=(12, 6))
for country in top_10_countries:
    country_data = top_10_data[top_10_data['Country'] == country]
    plt.plot(country_data['Year'], country_data['Happiness Score'], label=country)

plt.title("Top 10 Happiest Countries' Happiness Score Trend (2015-2022)")
plt.xlabel("Year")
plt.ylabel("Happiness Score")
plt.legend(title="Countries", loc='best')
plt.xticks(range(2015, 2023))
plt.show()
```

**2. Top 10 Happiest Countries in 2015 vs. 2022**

```python
# Ensure Happiness Score is numeric
whr['Happiness Score'] = pd.to_numeric(whr['Happiness Score'], errors='coerce')

# Create the Top 10 Comparison Bar Chart
top_10_2015 = whr[whr['Year'] == 2015].nlargest(10, 'Happiness Score')[['Country', 'Happiness Score']]
top_10_2022 = whr[whr['Year'] == 2022].nlargest(10, 'Happiness Score')[['Country', 'Happiness Score']]

top_10_comparison = pd.merge(top_10_2015, top_10_2022, on='Country', suffixes=('_2015', '_2022'))

# Plot the chart
plt.figure(figsize=(12, 6))
top_10_comparison.set_index('Country').plot(kind='bar', figsize=(10, 6))
plt.title("Top 10 Happiest Countries in 2015 vs. 2022")
plt.ylabel("Happiness Score")
plt.xticks(rotation=45)
plt.show()
```

**3. Ranking Changes Over Time (Heatmap)**

```python
# Get top 10 happiest countries in 2022
top_10_countries = whr[whr['Year'] == 2022].sort_values(by='Happiness Rank').head(10)['Country']
top_10_data = whr[whr['Country'].isin(top_10_countries)]
ranking_pivot_top_10 = top_10_data.pivot_table(index='Country', columns='Year', values='Happiness Rank')

# Plot the heatmap
plt.figure(figsize=(20, 7))
sns.heatmap(ranking_pivot_top_10, cmap="coolwarm", annot=True, fmt=".0f", cbar_kws={'label': 'Ranking'},
            linewidths=0.5, square=True)
plt.title("Top 10 Countries' Ranking Changes Over Time (2015-2022)")
plt.xlabel("Year")
plt.ylabel("Country")
plt.show()
```
